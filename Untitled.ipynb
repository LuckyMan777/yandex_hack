```python
import sqlite3
conn = sqlite3.connect('activities.db', isolation_level=None)
curr = conn.cursor()
```

```python
curr.executescript(q)
```

```python
curr.execute('select * from act')
```

```python
acts = curr.fetchall()
```

```python
acts
```

```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import NearestNeighbors
```

```python
df = pd.DataFrame(acts, columns=['act', 'compl', 'price'], dtype=int)
df.index = df['act'].values
df.drop(columns=['act'], inplace=True)
```

```python
df
```

```python
scaler = MinMaxScaler()
```

```python
df[['compl', 'price']] = scaler.fit_transform(df)
```

```python
df
```

```python
scaler.transform([[0, 100]])
```

```python
model_knn = NearestNeighbors(metric='manhattan', algorithm='brute', n_neighbors=100, n_jobs=-1)
```

```python
model_knn.fit(df)
```

```python
model_knn.kneighbors([df.loc['йога']], n_neighbors=len(df))
```

```python
indices = model_knn.kneighbors(scaler.transform([[0, 100]]), n_neighbors=46)[1][0]
```

```python
df.iloc[indices]
```
